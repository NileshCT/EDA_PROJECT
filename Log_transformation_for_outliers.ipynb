```python
#Below are the libraries that are used in order to perform EDA (Exploratory data analysis).
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import warnings
warnings.filterwarnings('ignore')
     
```

```python
#Import CSV file from google drive:
data=pd.read_csv("/content/drive/MyDrive/ds_salaries.csv")
```

```python
data.hist(figsize=(14,14), xrot=45)
plt.show()
```

```python
log_var=["remote_ratio","salary_in_usd","salary"]
fig=plt.figure(figsize=(24,10))
for j in range(len(log_var)):
  var=log_var[j]
  transformed="log_" + var
  data[transformed]=np.log10(data[var]+1)
  sub=fig.add_subplot(2,5,j+1)
  sub.set_xlabel(var)
  data[transformed].plot(kind="hist")
```
